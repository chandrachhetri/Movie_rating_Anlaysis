```python
#20. Classify Movies Based on Ratings [Excellent, Good, and Average]
data.columns
```

```python
def rating(rating):
    if rating >=8.0:
        return "Excellent"
    elif rating >=7.5:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

```python
#22.Find the unique values from Genre 
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
        list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```
